# KKBox Dataset Preprocessing

```python
# Load Python libraries
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

Importing KKBox data

```python
df_train = pd.read_csv('C:/Users/ekwaning/Documents/Flatiron/Capstone/Music-Recommendation/Data/train.csv')
df_train
```

```python
df_train = df_train.sample(frac=0.01)
df_train
```

```python
df_songs = pd.read_csv('C:/Users/ekwaning/Documents/Flatiron/Capstone/Music-Recommendation/Data/songs.csv')
df_songs.head(10)
```

```python
df_members = pd.read_csv('C:/Users/ekwaning/Documents/Flatiron/Capstone/Music-Recommendation/Data/members.csv')
df_members.head(10)
```

Merging Data sets

```python
#leftjoin songs data to main
df_train = pd.merge(df_train, df_songs, on='song_id', how='left')
df_train.head(10)
```

```python
#leftjoin members data to main
df_train = pd.merge(df_train, df_members, on='msno', how='left')
df_train.head(10)
```

```python
df_train.info()
```

Missing Data values

```python
# Count Na in %
df_train.isnull().sum()/df_train.isnull().count()*100
```

```python
# Replace NA
for i in df_train.select_dtypes(include=['object']).columns:
    df_train[i][df_train[i].isnull()] = 'unknown'
df_train = df_train.fillna(value=0)
```

```python
# Count Na in %
df_train.isnull().sum()/df_train.isnull().count()*100
```

Format Dates

```python
# Create Dates

# registration_init_time
df_train.registration_init_time = pd.to_datetime(df_train.registration_init_time, format='%Y%m%d', errors='ignore')
df_train['registration_init_time_year'] = df_train['registration_init_time'].dt.year
df_train['registration_init_time_month'] = df_train['registration_init_time'].dt.month
df_train['registration_init_time_day'] = df_train['registration_init_time'].dt.day

# expiration_date
df_train.expiration_date = pd.to_datetime(df_train.expiration_date,  format='%Y%m%d', errors='ignore')
df_train['expiration_date_year'] = df_train['expiration_date'].dt.year
df_train['expiration_date_month'] = df_train['expiration_date'].dt.month
df_train['expiration_date_day'] = df_train['expiration_date'].dt.day
```

```python
df_train.info()
```

```python
# Dates to category
df_train['registration_init_time'] = df_train['registration_init_time'].astype('category')
df_train['expiration_date'] = df_train['expiration_date'].astype('category')
```

```python
# Object data to category
for col in df_train.select_dtypes(include=['object']).columns:
    df_train[col] = df_train[col].astype('category')
    
# Encoding categorical features
for col in df_train.select_dtypes(include=['category']).columns:
    df_train[col] = df_train[col].cat.codes
```
